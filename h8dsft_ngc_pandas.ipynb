#Background

By looking at the columns in the dataset, i assume this table is used to document disease or pandemic cases on some countries in the data.

The table shows us the number of confirmed, deaths, and recovered cases in a pandemic, with the location of country, region, coordinates to track the origin of the data, and also the date of data collection.

After more exploration, this table only shows us number of cases in three countries: Malaysia, Singapore, and Indonesia.

Based on that, we want to see the statistics of disease cases in those 3 countries.

#Answer

##Ganti nama beberapa kolom

- Ganti nama country_region menjadi country saja, karena di kolom tersebut memang cuma ada country.

- Ganti nama location_geom jadi coordinates, karena itu merupakan gabungan dari latitude dan longitude.

```python
df2 = df2.rename(columns = {'country_region': 'country',
                            'location_geom': 'coordinates'})
df2
```

##Ganti value yang ada di kolom coordinate


```python
df2.drop(columns = 'coordinates', axis =1, inplace=True)
df2
```

```python
df2.insert(5, 'coordinates', df2['latitude'].astype(str) + ',' + df2['longitude'].astype(str))
df2
```

## Ganti tipe data

```python
df2.info()
```

```python
df2['date'] = pd.to_datetime(df2['date'])
df2.info()
```

## Drop beberapa kolom

Drop kolom yang null valuenya sampai 80% lebih dari data

```python
df2.drop(columns = ['fips', 'admin2'], axis =1, inplace=True)
df2
```

Drop kolom province_state dan combined_key karena yang punya nilai itu hanya negara Malaysia, sedangkan Indonesia dan Singapura tidak punya. Sulit untuk mengisi provinsi untuk kedua negara tersebut. combined_key sendiri adalah kode yang menggabungkan provinsi dengan negaranya.

```python
df2.drop(columns = ['province_state','combined_key'], axis=1, inplace=True)
df2
```

```python
df2.isna().sum()
```

```python
df2.drop(columns = ['recovered','active'], axis=1, inplace=True)
df2
```

## Drop baris yang kosong

```python
df2.isna().sum()
```

```python
df2.dropna(subset=['latitude'], inplace=True)
df2
```

```python
df2.isna().sum()
```

```python
df2.info()
```

#Conclusion

Based on the statistics, the total number of disease cases in Malaysia is 2,127,934 with confirmed deaths reaching 624,641. The rasio of confirmed and deaths cases in Malaysia seems to be off and could indicate an error in the data input process or other outliers

In Singapore, the number of cases is higher than Malaysia with 2,235,294 but the number of deaths is significantly lower with 1,722 confirmed deaths.

Lastly in Indonesia, the number of cases is the highest among the SEA countries with total confirmed cases reaching 6,738,225 and death toll of 160,941. This could be caused by Indonesia having the most population in the South-East Asia region with around 278 million people, compared to Singapore with only 5.4 million. Indonesia also ranked #4 in population globally.
